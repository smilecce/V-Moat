```python
# 本代码的主要作用就是生成一个类似原来的adv_statistic.csv的csv文件

import numpy as np
import pandas as pd
from glob import glob 
import os
from tqdm import tqdm
# os.environ["CUDA_VISIBLE_DEVICES"] = "0.1,2"
```

```python
adv_test_clean_path = '/home/vmoat/vmoat_code/user_study_adv_audio/attack_verification_test-clean'
all_clean_wav_name = [x.split('_')[4] for x in os.listdir(adv_test_clean_path)]
all_clean_wav_name
```

```python

cnt = 0
print(cnt)
temp_list = []
for idx,all_wav_folder in enumerate(tqdm(os.listdir(adv_test_clean_path))):
#     这个文件夹里面的wav文件，存放在下面的list中
#     print(all_wav_folder)
    this_folder_name = glob(f'/home/vmoat/vmoat_code/user_study_adv_audio/attack_verification_test-clean/{all_wav_folder}/*.wav')
#     print(len(this_folder_name)) #500个
    for wav_name in this_folder_name:
#         print(wav_name)
#         wav_name:/home/vmoat/vmoat_code/user_study_adv_audio/attack_verification_test-clean/model_weight_200_audi_8230-279154-0037_adv_x/wav_x_adv_i122_0.4490121490198645.wav
        wav_audibility2 = float(wav_name.split('/')[-1].split('_')[4][:-4])
#         表示这个是在这个循环的第几次
#         print(audibility2)
        wav_iter = wav_name.split('/')[-1].split('_')[3]
    
        wav_id = f'{all_clean_wav_name[idx]}-{wav_iter}'
#         print(wav_id) 8230-279154-0037-i122
        wav_id_list = all_clean_wav_name[idx].split('-')
        wav_ori_path = f'/home/vmoat/LibriSpeech/test-clean/{wav_id_list[0]}/{wav_id_list[1]}/{all_clean_wav_name[idx]}.flac'
        
        wav_adv_path = wav_name
                               
        temp_list.append([wav_id, wav_ori_path, wav_adv_path, wav_audibility2])                       
#         下面就开始创建csv文件                       
#         pd_data.loc[cnt,'id'] = wav_id
#         pd_data.loc[cnt,'ori_path'] = wav_ori_path
#         pd_data.loc[cnt,'adv_path'] = wav_adv_path
#         pd_data.loc[cnt,'audibility2'] = wav_audibility2
#         cnt = cnt+1
pd_data = pd.DataFrame(temp_list, columns=['id', 
                                    'ori_path', 
                                    'adv_path',
                                    'audibility2'])

```

```python
intervals = [0.03125, 0.0625, 0.125, 0.25, 0.5, 1, 2, 4, 8, 16, 32, 64]
print(type(pd_data['audibility2'][0]))
frames = []
for i in range(len(intervals)-1):
    frames.append(pd_data[(pd_data['audibility2']>=intervals[i]) & (pd_data['audibility2']<intervals[i+1])])
    print(len(frames[-1]))
    frames[-1].to_csv(f'/home/vmoat/userstudy_wsbnew/adv_statistics_11_level/adv_statistics_{i}.csv')
```

```python
clean_data = pd_data.drop_duplicates('ori_path', keep='first')
clean_data.index = range(len(clean_data))
# clean_data['adv_path'] = clean_data['ori_path']

for idx in clean_data.index:
    clean_data.loc[idx, 'adv_path'] = clean_data.loc[idx, 'ori_path']
#     clean_data.loc[idx, 'audibility1'] = 0
    clean_data.loc[idx, 'audibility2'] = 0
    clean_data.loc[idx, 'id'] = clean_data.loc[idx, 'id'].split('_')[0]+'_clean'

# clean_data
clean_data.to_csv('/home/vmoat/userstudy_wsbnew/statistics/clean_statistics.csv')
```
